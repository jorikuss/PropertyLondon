```python
import pandas as pd
pd.options.display.max_rows = 4000
```

```python
df = pd.read_csv("zoopla.csv")
```

```python
df.loc[:, "agent_main"] = df["agent"].str.split(" - ").map(lambda x: x[0])
df["agent_main"].value_counts()
df.loc[:, "number of images"] = df["number of images"].astype(str).str.split().map(lambda x: x[-1]).astype(float)
```
