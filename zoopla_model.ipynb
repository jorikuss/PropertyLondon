```python
import pandas as pd
```

```python
df = pd.read_csv("zoopla.csv")
df
```

```python
df[["bedrooms","bathrooms","receptions"]] = df[["bedrooms","bathrooms","receptions"]].fillna(0).astype('int32')
df["postcode"] = df["address"].str.split().str[-1]
```

```python
postcodes = list(pd.read_csv("postcodes.csv")["0"])
df = df[df['postcode'].isin(postcodes)]
```

```python
df['price'] = df['price'].str.replace(',', '').str.replace('£', '')
```

```python
df = df[df["price"] != "POA"]
df["price"] = df["price"].astype("int64")
```
